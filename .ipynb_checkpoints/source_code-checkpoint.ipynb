```python
#process signal channels
# List of EEG channels
eeg_channels = ['Delta_TP9', 'Delta_AF7', 'Delta_AF8', 'Delta_TP10',
                'Theta_TP9', 'Theta_AF7', 'Theta_AF8', 'Theta_TP10',
                'Alpha_TP9', 'Alpha_AF7', 'Alpha_AF8', 'Alpha_TP10',
                'Beta_TP9', 'Beta_AF7', 'Beta_AF8', 'Beta_TP10',
                'Gamma_TP9', 'Gamma_AF7', 'Gamma_AF8', 'Gamma_TP10']

df[eeg_channels].head()
```

```python
# Feature Extraction
```

```python
# Function for feature extraction
def extract_features(signal):
    features = []
    
    # Kurtosis
    kurtosis_value = signal.kurtosis()
    features.append(kurtosis_value)
    
    # 2nd Difference Mean
    second_diff = np.diff(signal, 2)
    second_diff_mean = np.mean(second_diff)
    features.append(second_diff_mean)
    
    # 2nd Difference Max
    second_diff_max = np.max(np.abs(second_diff))
    features.append(second_diff_max)
    
    # Mean, Std, Skewness
    mean_value = np.mean(signal)
    std_dev_value = np.std(signal)
    skewness_value = signal.skew()
    
    features.extend([mean_value, std_dev_value, skewness_value])
    
    return features

# Apply feature extraction to each row
extracted_features = df1[eeg_channels].apply(extract_features, axis=1)

# Create new column names
feature_names = ['Kurtosis', '2ndDiffMean', '2ndDiffMax', 'Mean', 'StdDev', 'Skewness']

# Create a new DataFrame with the extracted features
features_df = pd.DataFrame(extracted_features.tolist(), columns=feature_names)

# Concatenate the new features DataFrame with the existing DataFrame
result_df = pd.concat([df1, features_df], axis=1)

# Display the result DataFrame with the extracted features
result_df
```

```python
# Scaling
```

```python
# Extract the features (excluding the 'Class' column, assuming it contains labels)
features = result_df.drop(['label','TimeStamp'], axis=1)

# Initialize the StandardScaler with centering (zero-mean scaling)
scaler = StandardScaler(with_mean=True, with_std=False)

# Fit and transform the features using the scaler
scaled_features = scaler.fit_transform(features)

# Create a new DataFrame with the zero-mean scaled features
scaled_df = pd.DataFrame(scaled_features, columns=features.columns)

# Add the 'Class' column back to the zero-mean scaled DataFrame
scaled_df['label'] = result_df['label']

# Display the zero-mean scaled DataFrame
scaled_df
```

# Variance Thresholding Feature Selection


```python
# Separate features and labels
X = scaled_df.drop('label', axis=1)  # Features
y = scaled_df['label']  # Labels

# Apply Variance Thresholding feature selection
variance_threshold = 0.1  # Adjust the threshold as needed
variance_selector = VarianceThreshold(threshold=variance_threshold)
X_selected_variance = variance_selector.fit_transform(X)

# Create a new DataFrame with selected features
selected_features_variance_df = pd.DataFrame(data=X_selected_variance, columns=X.columns[variance_selector.get_support()])

print("Selected Features: ")
print(selected_features_variance_df.columns)
# Add the 'Class' column back to the selected features DataFrame
selected_features_variance_df['label'] = y
```

```python
# Selected Features
```

```python
import pandas as pd
from sklearn.model_selection import cross_val_score, train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, confusion_matrix

# Assuming the last column is the target label 'label'
X = selected_features_variance_df.drop('label', axis=1)  # Features
y = selected_features_variance_df['label']  # Target label

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Classifiers
knn_classifier = KNeighborsClassifier()
svm_classifier = SVC()
mlp_classifier = MLPClassifier()
nb_classifier = GaussianNB()
rf_classifier = RandomForestClassifier()
lda_classifier = LinearDiscriminantAnalysis()

classifiers = {
    'KNN': knn_classifier,
    'SVM': svm_classifier,
    'MLP': mlp_classifier,
    'NB': nb_classifier,
    'RF': rf_classifier,
    'LDA': lda_classifier,
}

# Evaluate each classifier
for clf_name, clf in classifiers.items():
    # 10-fold cross-validation
    scores = cross_val_score(clf, X_train, y_train, cv=10, scoring='accuracy')
    print(f"{clf_name} Cross-Validation Accuracy: {scores.mean():.4f}")

    # Train the classifier
    clf.fit(X_train, y_train)

    # Make predictions on the test set
    y_pred = clf.predict(X_test)

    # Evaluate performance on the test set
    accuracy = accuracy_score(y_test, y_pred)
    confusion_mat = confusion_matrix(y_test, y_pred)

    sensitivity = confusion_mat[0, 0] / (confusion_mat[0, 0] + confusion_mat[0, 1])
    specificity = confusion_mat[1, 1] / (confusion_mat[1, 0] + confusion_mat[1, 1])

    print(f"{clf_name} Test Accuracy: {accuracy:.4f}")
    print(f"{clf_name} Test Sensitivity: {sensitivity:.4f}")
    print(f"{clf_name} Test Specificity: {specificity:.4f}\n")
```
